## Choosing an activation function

In some senses, the choice of activation function is arbitrary. There is little literature that suggests that one activation is better in all cases than the rest. Recent work in the ML space has shown Sigmoid Linear Units (SiLUs) to be popular, activation functions can be swapped out to see which performs best during training


First, import the modules used in this notebook.

```python
import torch
import matplotlib.pyplot as plt
import numpy as np
```

## Common Activation functions

### ReLu
One of the simplest and most commonly used activation functions, a Rectified Linear Unit (or ReLU) passes all positive values through unchanged, while zeroing out any negative values. This function acts to supress any signals that are not considered important (a negative value), while preserving values that are (positive ones).



```python
#create the "x values"
relu_input = torch.linspace(-5, 5, 15)

#create the output from the relu, or the "y values"
relu_output = torch.relu(relu_input)

plt.plot(relu_input, relu_output)
plt.xlabel("input to ReLU layer")
plt.ylabel("output from ReLU layer")

plt.plot()
```

## Leakey ReLU

A leakey ReLU is a relu that allows a small amount of the negative component of the signal to "leak" through. Leakey ReLUs attempt to address some tricky machine learning problems like vanishing or exploding gradients.

```python
#create the "x values"
leakey_relu_input = torch.linspace(-5, 5, 50)

#create the output from the relu, or the "y values"



alpha_values = np.array([.1, 0.3, 0.5, 1], np.float32)

for plot_idx in range(4):
        leakey_relu = torch.nn.ELU(alpha= alpha_values[plot_idx], inplace=False)
        leakey_relu_output = leakey_relu(leakey_relu_input)
        plt.subplot(2, 2, plot_idx+1)
        plt.xlabel("ELU input")
        plt.ylabel("ELU output")
        plt.plot(leakey_relu_input, leakey_relu_output)

# plt.plot(leakey_relu_input, leakey_relu_output)
# plt.xlabel("input to leakeyReLU layer")
# plt.ylabel("output from leakeyReLU layer")

# plt.plot()
```

## Elu

A variation on the Relu is the Exponential Linear Unit, or Elu. Elus have a smoother response, and can be parameterized to have different cutoff knees. 
[consult here for the paper on Elus](https://arxiv.org/abs/1511.07289)

```python
elu_input = torch.linspace(-5, 5, 100)
elu_activation_function = torch.nn.ELU(alpha = .001,  inplace=False)


alpha_values = np.array([0.1, 0.3, 0.7, 1], np.float32)

for plot_idx in range(4):
        elu_activation_function = torch.nn.ELU(alpha= alpha_values[plot_idx], inplace=False)
        elu_output = elu_activation_function(elu_input)
        plt.subplot(2, 2, plot_idx+1)
        plt.xlabel("ELU input")
        plt.ylabel("ELU output")
        plt.plot(elu_input, elu_output)

```

### Sigmoid
Sigmoids are functions that push values to 0 and 1. Sigmoids are useful for final classification in order to make a yes/no decision, like you would want for a classification.

```python
#create the "x values"
sigmoid_input = torch.linspace(-5, 5, 20)

#create the output from the relu, or the "y values"
sigmoid_output = torch.sigmoid(sigmoid_input)

plt.plot(sigmoid_input, sigmoid_output)
plt.xlabel("input to relu layer")
plt.ylabel("output from relu layer")

plt.plot()
```

## SiLU

A Sigmoid Linear Unit (SiLU) is like a combination of a ReLU and a Sigmoid. Like ReLUs, SiLUs don't saturate to 1, but they have a smoother curve and give a near-zero output at zero input. SiLUs are popular in the literature as of the writing of this notebook. Notice how there is a dip into negative outputs right before x=0, giving it a knee shape reminiscent of Infinite Impulse Response (IIR) filters in digital signal processing.

```python
#create the "x values"
silu_input = torch.linspace(-7, 3, 20)

#create the output from the relu, or the "y values"
silu_output = torch.nn.SiLU()(silu_input)

plt.plot(silu_input, silu_output)
plt.xlabel("input to SiLU layer")
plt.ylabel("output from SiLU layer")

plt.plot()
```
